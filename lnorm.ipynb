```python
import numpy as np
import math
import matplotlib.pyplot as plt


def lnorm(mu, N, mid=400):
    # - infinity < mu < math.log(400)
    sigma = math.sqrt((math.log(mid)-mu)*2)
    return np.random.lognormal(mu, sigma, N)
```

```python
s = lnorm(5, 10000, 300)
plt.hist(s+1000, bins=200)
plt.show()
```

```python
s = lnorm(math.log(1000)/1.05, 10000,1000)
plt.hist(s, bins=200)
plt.show()
```

```python
s = lnorm(5.6, 1000)
print(sum(s)/len(s))
```

```python
mid = 500
low = 200
n = 100

assert low < mid
mu = 5.8
sigma = math.sqrt((math.log(mid-low) - mu) * 2)
r = np.random.lognormal(mu, sigma, n)+low
plt.hist(r, bins=200)
plt.show()
```

```python
def get_lognorm(low, mid, n):
    assert low < mid
    mu = math.log(mid-low)/1.05
    sigma = math.sqrt((math.log(mid-low) - mu) * 2)
    return np.random.lognormal(mu, sigma, n)+low

r = get_lognorm(2, 5, 10000)

plt.hist(r, bins=200)
plt.show()
print(sum(r)/len(r))
```
